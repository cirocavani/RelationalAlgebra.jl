Find the names of all people who frequent every pizzeria serving at least one pizza they eat.

```julia
include("pizza_data.jl")
```

```julia
r1 = Eats ⨝ Serves
```

```julia
r2 = π(r1, :name, :pizzeria)
```

```julia
r3 = r2 - Frequents
```

```julia
r4 = π(r3, :name)
```

```julia
r5 = π(Person, :name)
```

```julia
r6 = r5 - r4
```

```julia
result = [t[1] for t in r6.tuples_values]
@assert length(result) == 1
@assert "Fay" ∈ result
```
